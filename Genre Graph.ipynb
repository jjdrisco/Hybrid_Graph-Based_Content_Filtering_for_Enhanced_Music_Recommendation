```python
import networkx as nx
from pyvis.network import Network

# Consolidated genres
genre_groups = {
    'aggressive-fusion': ['dubstep', 'grunge', 'metal'],
    'industrial': ['goth', 'heavy-metal', 'industrial'],
    'punk-rock': ['alt-rock', 'garage', 'hard-rock', 'j-rock', 'punk', 'punk-rock'],
    'hardstyle': ['happy', 'hardstyle'],
    'disco-ska': ['disco', 'ska', 'synth-pop'],
    'rock': ['alternative', 'rock'],
    'anime': ['anime', 'club'],
    'edm-house': ['deep-house', 'electronic', 'progressive-house'],
    'edm': ['dub', 'edm', 'electro', 'groove', 'house'],
    'j-dance': ['dancehall', 'j-dance'],
    'funk-hip-hop': ['funk', 'hip-hop'],
    'latin': ['dance', 'latin', 'latino', 'reggae', 'reggaeton'],
    'pop': ['k-pop', 'pop', 'pop-film'],
    'brazilian': ['brazil', 'mpb'],
    'blues-rnb': ['blues', 'j-pop', 'r-n-b'],
    'indie': ['folk', 'indie', 'indie-pop', 'psych-rock'],
    'chill': ['chill', 'sad'],
    'pagode-samba': ['pagode', 'samba', 'sertanejo'],
    'country-soul': ['country', 'soul'],
    'rock-n-roll': ['rock-n-roll', 'rockabilly'],
    'chicago-house': ['chicago-house', 'detroit-techno'],
    'jazz-tango': ['honky-tonk', 'jazz', 'tango'],
    'vocal-pop': ['acoustic', 'cantopop', 'mandopop', 'singer-songwriter', 'songwriter'],
    'disney': ['disney', 'guitar'],
    'soundscape': ['ambient', 'new-age']
}

# Define descriptive edges with influences
edges = [
    ('blues', 'jazz', 'primary origin'), ('jazz', 'soul', 'primary origin'),
    ('soul', 'r-n-b', 'primary origin'), ('r-n-b', 'hip-hop', 'primary origin'),
    ('hip-hop', 'rap', 'primary origin'), ('country', 'rock-n-roll', 'primary origin'),
    ('rock-n-roll', 'rock', 'primary origin'), ('rock', 'punk', 'primary origin'),
    ('punk', 'hardcore', 'primary origin'), ('hardcore', 'metalcore', 'primary origin'),
    ('metalcore', 'metal', 'primary origin'), ('metal', 'black-metal', 'primary origin'),
    ('electronic', 'edm', 'primary origin'), ('edm', 'house', 'primary origin'),
    ('house', 'deep-house', 'primary origin'), ('house', 'techno', 'primary origin'),
    ('techno', 'trance', 'primary origin'), ('trance', 'dubstep', 'primary origin'),
    ('dubstep', 'drum-and-bass', 'primary origin'), ('reggae', 'dancehall', 'primary origin'),
    ('dancehall', 'afrobeat', 'primary origin'), ('folk', 'bluegrass', 'primary origin'),
    ('pop', 'dance', 'primary origin'), ('pop', 'electro', 'primary origin'),
    ('electro', 'synth-pop', 'primary origin'), ('synth-pop', 'new-age', 'primary origin'),
    ('j-pop', 'j-rock', 'primary origin'), ('k-pop', 'pop', 'primary origin'),
    ('latin', 'salsa', 'primary origin'), ('salsa', 'samba', 'primary origin'),
    ('classical', 'opera', 'primary origin'), ('opera', 'show-tunes', 'primary origin'),
    ('alternative', 'alt-rock', 'primary origin'), ('alt-rock', 'grunge', 'primary origin'),
    ('grunge', 'psych-rock', 'primary origin'), ('punk-rock', 'punk', 'primary origin')
]

# Create Graph
G = nx.Graph()
for group, subgenres in genre_groups.items():
    for genre in subgenres:
        G.add_node(genre, group=group)
        G.add_edge(group, genre, label='subgenre')
for edge in edges:
    G.add_edge(edge[0], edge[1], label=edge[2])

# Visualization
net = Network(notebook=True, height='750px', width='100%', bgcolor='#222222', font_color='white')
net.from_nx(G)
net.show("music_genre_graph.html")
```

```python
shortest_paths = nx.single_source_shortest_path(G, source="hip-hop")

for genre, path in shortest_paths.items():
    print(f"Shortest path from Hip Hop to {genre}: {path}")
    
pick = np.random.choice(df["track_genre"].unique())
    
shortest_paths[pick]
```

```python
from collections import Counter

# Get all shortest paths from Rap
shortest_paths = nx.single_source_shortest_path(G, source="hip-hop")

# Collect second genres (the genre right after "Rap" in each path)
second_genres = [path[1] for path in shortest_paths.values() if len(path) > 1]

# Count occurrences of each second genre
genre_counts = Counter(second_genres)

# Find the most common second genre
most_common_second_genre, count = genre_counts.most_common(1)[0]

print(f"The most common second genre after 'Rap' is: {most_common_second_genre} (appears {count} times)")
```
